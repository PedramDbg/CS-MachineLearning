```python
# apply Counter on data 
counter = Counter(str(data['test.txt']))
#print(data.keys())
#print(data['test.txt'])
#print(type(counter))
```

```python
import re
# make all the characters lowercase
data_cleaned = str(data['test.txt']).lower()
print(type(data_cleaned))
# Remove anything that isn't a word character or a space
# We could use .replace(".", "") but regex is a lot easier!

data_cleaned = re.sub("[^\w ]", "", data_cleaned)

print("Cleaned text is:", data_cleaned)
```

```python
import string
alphabet = list(string.ascii_lowercase)
print(alphabet)
```

```python
freq = []
for i in alphabet:
    freq = np.append(freq, counter[i])
    
freq_scaled = freq/np.max(freq)
print(freq)
```

```python
dics = {}

for i, let in enumerate(alphabet):
    dics[let] = freq[i]
print(dics)
```

```python
x = np.arange(len(alphabet))

fig, axs = plt.subplots(2, figsize=(20,16))
axs[0].bar(x, height=freq_scaled)
axs[0].set_xticks(x, alphabet)
axs[0].set_title('Normalized data', fontsize='24')
axs[0].set_xlabel('Alphabets', fontsize='18')
axs[0].set_ylabel('Frequency', fontsize='18')

axs[1].bar(x, height=freq)
axs[1].set_xticks(x, alphabet)
axs[1].set_title('Unnormalized data', fontsize='24')
axs[1].set_xlabel('Alphabets', fontsize='18')
axs[1].set_ylabel('Frequency', fontsize='18')
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('test.png')
```

```python
p1 = np.array([255, 0, 0])
p2 = np.array([0, 255, 0])
p3 = np.array([0, 0, 255])
p = [p1, p2, p3]
```

```python
img_1 = np.copy(img)
for i in range(np.shape(img_1)[0]):
  for j in range(np.shape(img_1)[1]):
    mag = np.zeros(np.shape(p)[0])
    for k in range(np.shape(p)[0]):
      diff = abs(img_1[i,j] - p[k])
      mag[k] = np.linalg.norm(diff)
      #mag[k] = magnitude(diff)
    ind_min = np.argmin(mag)
    img_1[i,j] = np.copy(p[ind_min])
```

In this part of the assignment, you will be using NumPy to solve the following word problem. Specifically, you can use the LinAlg package from Numpy to solve the system of equations.

Problem statement:
You and a friend go to buy tacos. You get three soft tacos
and three burritos and your total bill is $11.25. Your
friend's bill is $10.00 for four soft tacos and two
burritos. How much do soft tacos cost? How much do burritos
cost?

```python
import numpy as np
import numpy.linalg as LA
a = np.array([[3, 3], [4, 2]])
b = np.array([11.25, 10.00])
x = np.linalg.solve(a, b)
print('soft taco costs' , x[0] , 'and', 'burrito costs', x[1])
```
